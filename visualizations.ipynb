```
from datetime import timedelta, datetime
import math
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, FloatType, TimestampType
from pyspark.sql.functions import col,current_date, current_timestamp, date_format,count,avg,col,asc,desc

df2 = spark.read.option("inferSchema" , "true").parquet(f'/mnt/bucket-big-traffic/big_data_prj/demo_example/big')

gold_car = (spark.read
                .option("inferSchema","true")
                .table("default.gold_car"))

gold_sound = (spark.read
                .option("inferSchema","true")
                .table("default.gold_sound"))

historical_car = (spark.read
                .option("inferSchema","true")
                .table("default.historical_car"))

historical_sound = (spark.read
                .option("inferSchema","true")
                .table("default.historical_sound"))


df3=gold_car.withColumn("timestamp_hourly", date_format(col("time"), "yyyy-MM-dd HH:00"))

hourly_aggregated=df3.groupby(['city','timestamp_hourly','lat','lon']).agg({'reg_no':'count'})
df4 = historical_car.withColumn("timestamp_hourly", date_format(col("time"), "yyyy-MM-dd HH:00"))

historical_hourly_aggregated=df4.groupby(['city','timestamp_hourly','lat','lon']).agg({'reg_no':'count'})
    
display(hourly_aggregated)

display(gold_car)

display(gold_sound)

display(historical_sound)

display(historical_hourly_aggregated)

display(gold_car.groupBy("city") 
    .agg(count("reg_no").alias("count_cars")) 
    .orderBy(col("count_cars").asc())
    .take(5))

display(gold_sound.groupBy("city") 
    .agg(avg("decibels").alias("avg_decibels"))
    .orderBy(col("avg_decibels").asc())
    .take(5))

display(gold_car.groupBy("city") 
    .agg(count("reg_no").alias("count_cars")) 
    .orderBy(col("count_cars").desc())
    .take(5))

display(gold_sound.groupBy("city") 
    .agg(avg("decibels").alias("avg_decibels"))
    .orderBy(col("avg_decibels").desc())
    .take(5))
```
